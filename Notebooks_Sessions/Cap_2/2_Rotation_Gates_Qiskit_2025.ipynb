```python
from math import pi

from qiskit import QuantumCircuit, QuantumRegister
from qiskit.quantum_info import Operator
from qiskit.visualization import array_to_latex
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_distribution
from qiskit import transpile
```

```python
# Descomente la siguiente linea si desea ver las imágenes más grandes.
%config InlineBackend.figure_format = 'svg' # Permite mejor calidad en las imágenes generadas.
```

```python
q = QuantumRegister(1, name='q')
```

### U gate:

$$
U(\theta, \phi, \lambda) = \begin{pmatrix}
\cos(\theta/2) & -e^{i\lambda}\sin(\theta/2) \\
e^{i\phi}\sin(\theta/2) & e^{i(\lambda+\phi)}\cos(\theta/2) 
\end{pmatrix}.

$$


```python
qc = QuantumCircuit(q)
qc.u(pi/2,pi/2,pi/2,q)
```

```python
qc.draw('mpl')
# qc.draw('mpl', style='iqx', filename='Ugate.png')
```

```python
unitary = Operator(qc)

array_to_latex(unitary, prefix="\\textit{U(}{\\pi}/2,{\\pi}/2,{\\pi}/2) =\n")
```

### P gate:

The $P(\lambda)= U(0, 0, \lambda)$ gate has the matrix form

$$
P(\lambda) = 
\begin{pmatrix}
1 & 0 \\
0 & e^{i \lambda}
\end{pmatrix},
$$

```python
qc = QuantumCircuit(q)

qc.p(pi/2, q)

qc.draw('mpl')
# qc.draw('mpl', style='iqx', filename='Pgate.png')
```

```python
unitary = Operator(qc)

array_to_latex(unitary, prefix="\\textit{U(0, 0, }{\\pi}/2) \\textit{= P(}{\\pi}/2) =\n")
```

### Standard Rotations

The standard rotation gates are those that define rotations around the Paulis $\hat{P}=\{X,Y,Z\}$. They are defined as 

$$ R_{\hat{P}}(\theta) = \exp(-i \theta \hat{P}/2) = \cos(\theta/2)I -i \sin(\theta/2)\hat{P}$$

#### Rotation around X-axis:

$$
R_x(\theta) = 
\begin{pmatrix}
\cos(\theta/2) & -i\sin(\theta/2)\\
-i\sin(\theta/2) & \cos(\theta/2)
\end{pmatrix} = U(\theta, -\pi/2,\pi/2)
$$

```python
qc = QuantumCircuit(q)
qc.rx(pi/2,0)

qc.draw('mpl')
#qc.draw('mpl', style='iqx', filename='RXGate.png')
```

```python
unitary = Operator(qc)
array_to_latex(unitary, prefix="\\textit{Rx(}{\\pi}/2) = \n")
```

#### Rotation around Y-axis:

$$
R_y(\theta) = 
\begin{pmatrix}
\cos(\theta/2) & \sin(\theta/2)\\
\sin(\theta/2) & \cos(\theta/2)
\end{pmatrix}  = U(\theta, 0, 0)
$$

```python
qc = QuantumCircuit(q)
qc.ry(pi/2,0)

qc.draw('mpl')
#qc.draw('mpl', style='iqx', filename='RYGate.png')
```

```python
unitary = Operator(qc)
array_to_latex(unitary, prefix="\\textit{Ry(}{\\pi}/2) = \n")
```

#### Rotation around Z-axis

$$
R_z(\phi) = 
\begin{pmatrix}
e^{-i \phi/2} & 0 \\
0 & e^{i \phi/2}
\end{pmatrix}\equiv P(\phi)
$$

```python
qc = QuantumCircuit(q)
qc.rz(pi/2,0)

qc.draw('mpl')
# qc.draw('mpl', style='iqx', filename='RZGate.png')
```

```python
unitary = Operator(qc)
array_to_latex(unitary, prefix="\\textit{Rz(}{\\pi}/2) = \n")
```
